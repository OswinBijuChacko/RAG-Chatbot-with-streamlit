```python
import os
import pandas as pd
from langchain_community.vectorstores import FAISS
from langchain.schema import Document
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
# Load the CSV file into a DataFrame and creating a document
df = pd.read_csv("knowledge_base.csv")
documents = [Document(page_content=row['Answer']) for _, row in df.iterrows()]
```

```python
# Initialize the HuggingFace embedding function
embedding_function = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# Extract raw text from documents
texts = [doc.page_content for doc in documents]

# Create FAISS vector store directly from texts using the embedding function as vectors
vector_store = FAISS.from_texts(texts=texts, embedding=embedding_function)
```

```python
def print_embeddings():
    embeddings = embedding_function.embed_documents(texts)
    for i, embedding in enumerate(embeddings):
        print(f"Document {i+1} Embedding: {embedding}")

# Example usage of print_embeddings
print_embeddings()
```
